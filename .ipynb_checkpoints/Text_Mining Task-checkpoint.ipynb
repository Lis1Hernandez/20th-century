```python
# important lib.s
from textblob import TextBlob
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
#import matplotlab
import nltk
#import nltk
from nltk.corpus import stopwords
import re
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab')
#nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
from collections import Counter
sns.set()
```

```python
# Import txt file
with open('key_event_20th_century.txt', 'r', errors='ignore') as file: 
   data = file.read().replace( '\n', ' ')
```

```python
# word tokenization 
from nltk.tokenize import word_tokenize
tokenized_word = word_tokenize(data)
```

```python
from nltk.probability import FreqDist
dist_words = FreqDist(tokenized_word)
```

```python
# saving top 10 most common
dist_words_top10 = dist_words.most_common(10)
# turning into series via dict to easier ploting
top_10_series = pd.Series(dict(dist_words_top10)) 

# using sns to creat ebar plot
fig, ax = plt.subplots(figsize=(10,10))
top_10_plot = sns.barplot(x=top_10_series.index, y=top_10_series.values, ax=ax)
```

```python
# Removing stopwords in words
# Create array of stop_words
stop_words = set(stopwords.words('english'))
stop_words.add("The")
filtered_words = [] 
for word in tokenized_word:
      if word not in stop_words:
           filtered_words.append(word)
# Substitute all punctuation marks with a space
sans_punc = re.sub("[^a-zA-Z]", " ", str(filtered_words))
```

```python
tokenized_word_filtered = word_tokenize(sans_punc)
dist_words_filtered = FreqDist(tokenized_word_filtered)

# re-run plot
dist_words_filtered_top10 = dist_words_filtered.most_common(10)
# turning into series via dict to easier ploting
top_10_filtered_series = pd.Series(dict(dist_words_filtered_top10)) 
fig, ax = plt.subplots(figsize=(10,10))
top_10_filtered_plot = sns.barplot(x=top_10_filtered_series.index, y=top_10_filtered_series.values, ax=ax)
```

What has changed? Comment on your impressions in a markdown cell under the plot. 

Removed unnecessary words and punctuation marks. For example, period and comma. This allowed us to accurately see the frequency. 

```python
# created tag list using text blob
text = TextBlob(str(tokenized_word_filtered))
tags_list = text.tags
```

```python
# creating df using tags
df_text = pd.DataFrame(tags_list)
df_text.columns = [ 'Words', "Word type"]
df_t=df_text.groupby('Word type').count().reset_index()
top10=df_t.nlargest(10, 'Words')
```

```python
# create df of POS tags
df_pos = df_text[df_text["Word type"] == "POS"]
df_pos.columns = ["Word", "Occurrences"]
x = df_pos.groupby('Word').count().reset_index()
y = x.sort_values(by = ['Occurrences'], ascending=False)
top10 = y.nlargest(10, 'Occurrences') 
# create plot
plt.figure(figsize=(10, 5))
with sns.dark_palette("xkcd:blue", 10):
      sns.barplot(x="Word", y="Occurrences",
      saturation=0.9, data = top10).set_title("20th Century Key Events - most frequently used part of speech")
```

```python
# nouns
df_nouns = df_text[(df_text['Word type'] == "NN") | (df_text['Word type'] == "NNS") | (df_text['Word type'] == "NNS")]
df_nouns.columns = ["Word", "Occurrences"]
x = df_nouns.groupby('Word').count().reset_index()
y = x.sort_values(by = ['Occurrences'], ascending=False)
top15 = y.nlargest(15, 'Occurrences') 
# create plot
plt.figure(figsize=(10, 5))
with sns.dark_palette("xkcd:blue", 10):
      sns.barplot(x="Word", y="Occurrences",
      saturation=0.9, data = top15).set_title("20th Century Key Events - most frequently used nouns")
```

```python
# verbs
df_verbs = df_text[(df_text['Word type'] == "VBD") | (df_text['Word type'] == "VBZ") | (df_text['Word type'] == "VBG")]
df_verbs.columns = ["Word", "Occurrences"]
x = df_verbs.groupby('Word').count().reset_index()
y = x.sort_values(by = ['Occurrences'], ascending=False)
top15 = y.nlargest(15, 'Occurrences') 
# create plot
plt.figure(figsize=(10, 5))
with sns.dark_palette("xkcd:blue", 10):
      sns.barplot(x="Word", y="Occurrences",
      saturation=0.9, data = top15).set_title("20th Century Key Events - most frequently used verbs")
```

```python
# adj
df_adj = df_text[(df_text['Word type'] == "JJ")]
df_adj.columns = ["Word", "Occurrences"]
x = df_adj.groupby('Word').count().reset_index()
y = x.sort_values(by = ['Occurrences'], ascending=False)
top15 = y.nlargest(15, 'Occurrences') 
# create plot
plt.figure(figsize=(10, 5))
with sns.dark_palette("xkcd:blue", 10):
      sns.barplot(x="Word", y="Occurrences", saturation=0.9, data = top15).set_title("20th Century Key Events - most frequently used adjectives")
```

```python
# create df of countries to turn into list of country
df_all_countries = pd.read_csv("countries_list_20th_century_1.5.csv")
# use lambda function to shorten country name to alias
df_all_countries['country_alias'] = df_all_countries['country_name'].apply(lambda x: x.rsplit(',',1)[0])
# remove spaces
df_all_countries['country_alias'] = df_all_countries['country_alias'].apply(lambda x: x.replace(" ",""))
# create list to use for search
countries_list = df_all_countries["country_alias"].to_list()
```

```python
# include only country names for count
country_filter = []
for word in filtered_words:
    if word in countries_list:
        country_filter.append(word)
```

```python
# count then split into different list to be the different columns for the df
country_count = Counter(country_filter)
dict_of_counts = {d: country_count[d] for d in country_filter}
country_list = dict_of_counts.keys()
count_list = dict_of_counts.values()

df_countries = pd.DataFrame({"Country":country_list, "Count":count_list})
```

```python
# turn countries df into bar plot
plt.figure(figsize = (8, 12))
with sns.dark_palette("xkcd:blue", 20):
    sns.barplot(x = "Count", y = "Country", orient = "h", saturation = 0.9, data = df_countries).set_title("20th Century Key Events - Country Count")
```

Is there anything in particular you notice in this plot? 

Germany, Japan and France are the most frequent Countries for 20th Century Events. With Germany having the most
significant frequency at approximately 45.
